```python
data_1_orders_2=  data_1_orders.rename(columns={"total_orders": "total_orders_deliverando"})
data_1_orders_2 = data_1_orders_2.drop(['name', 'kpi','Month 1','Month 2'], axis=1)
data_1_orders_2= data_1_orders_2.astype({'total_orders_deliverando': int },copy=False)
data_1_orders_2
```

```python
total_deliverando_orders= data_1_orders['total_orders'].sum()
total_deliverando_orders
```

```python
data_1_orders_2_zip_unique= data_1_orders_2['zip'].unique().tolist()
data_1_orders_2_zip_unique
```

```python
data_1_orders_3= data_1_orders_2.groupby(['zip','delivery service']).agg('sum')
data_1_orders_3= data_1_orders_3.rename(columns={"total_orders_deliverando": "total_orders"})
data_1_orders_3= data_1_orders_3.reset_index()
data_1_orders_3
```

```python
data_1_orders_3['%_from_total_orders']= round((data_1_orders_3['total_orders'] / total_deliverando_orders)* 100, 2)
data_1_orders_3
```

```python
competitors_data_2= competitors_data.rename(columns={"total_orders": "total_orders_competitors"})
competitors_data_2= competitors_data_2.drop(['name','orders_x','month_x','delivery_service_x','orders_y','month_y'], axis=1)
competitors_data_2= competitors_data_2.rename(columns={"delivery_service_y": "delivery service"})
competitors_data_2
```

```python
total_competitors_orders= competitors_data_2['total_orders_competitors'].sum()
total_competitors_orders
```

```python
competitors_data_2['%_from_total_orders']= round((competitors_data_2['total_orders_competitors'] / total_competitors_orders)* 100, 2)
```

```python
competitors_data_2['%_from_total_orders'].sum()
```

```python
len(competitors_data_2['zip'].unique())
```

```python
competitors_data_3= competitors_data_2[['zip', 'total_orders_competitors','%_from_total_orders']].groupby(['zip']).agg('sum')
competitors_data_3= competitors_data_3.rename(columns={"total_orders_competitors": "total_orders"})
competitors_data_3['delivery service']= 'competition'
competitors_data_3= competitors_data_3.reset_index()
competitors_data_3
```

```python
competitors_zip_ranked= competitors_data_3.sort_values(by=['total_orders'], ascending= False)
competitors_zip_ranked
```
